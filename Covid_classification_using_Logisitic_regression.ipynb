```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('Covid Data.csv')
```

```python
df['CLASIFFICATION_FINAL'].shape
```

```python
df=df.dropna()
```

```python
df['CLASIFFICATION_FINAL'].replace([4,5,6,7],0,inplace = True)
```

```python
df['CLASIFFICATION_FINAL'].replace([2,3],1,inplace = True)
```

```python
df = df.drop('DATE_DIED',axis =1)
```

```python
plt.figure(figsize=(15,15))
plot_num = 1

for i in df.columns:
  if plot_num <= 19:
    plt.subplot(6,6,plot_num)
    sns.boxplot(df[i])
  plot_num +=1
plt.tight_layout()
plt.show()
```

```python
crr = df.corr()
```

```python
plt.figure(figsize= (15,15))
sns.heatmap(crr,annot = True)
```

```python
X = df.drop(['CLASIFFICATION_FINAL','MEDICAL_UNIT','AGE','USMER','HIPERTENSION','INMSUPR'],axis =1 )
Y  = df['CLASIFFICATION_FINAL']
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train ,y_test = train_test_split(X,Y,test_size = .2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error,mean_absolute_error
accuracy_score(y_pred,y_test)
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
asd = mean_squared_error(y_pred,y_test)
asd
```

```python
rmse = np.sqrt(asd)
rmse
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier(criterion = 'gini')
clf1.fit(x_train,y_train)
```

```python
y_p1 = clf1.predict(x_test)
```

```python
accuracy_score(y_p1,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf2 = RandomForestClassifier(n_estimators=150, max_features=9, criterion='entropy' )
clf2.fit(x_train,y_train)
```

```python
y_p12 = clf2.predict(x_test)
```

```python
accuracy_score(y_p12,y_test)
```

```python
from sklearn.svm import SVC
clf3 = SVC(kernel = 'rbf')
clf3.fit(x_train,y_train)
```
